# The following is for trying to get some insights from clustering this data

```python
data_for_clustering = df[['V1', 'V2', 'Class']]

from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=2, random_state=42)
clusters = kmeans.fit_predict(data_for_clustering[['V1', 'V2']])

data_for_clustering['Cluster'] = clusters
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(data_for_clustering['V1'], data_for_clustering['V2'], c=data_for_clustering['Cluster'], cmap='viridis', alpha=0.5)
plt.title('Clustering Visualization using V1 and V2')
plt.xlabel('V1')
plt.ylabel('V2')
plt.colorbar(label='Cluster')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Cluster', hue='Class', data=data_for_clustering)
plt.title('Fraud Distribution Across Clusters')
plt.xlabel('Cluster')
plt.ylabel('Count')
plt.show()
```

```python
def time_of_day(hour):
    if 5 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Afternoon'
    elif 17 <= hour < 21:
        return 'Evening'
    else:
        return 'Night'

important['Time_of_Day'] = important['Hours'].apply(time_of_day)
fraud_df['Time_of_Day'] = important['Hours'].apply(time_of_day)
non_fraud_df['Time_of_Day'] = important['Hours'].apply(time_of_day)
```

```python
fraud_df['Time_of_Day'].value_counts()
```

```python
time_of_day_counts = fraud_df['Time_of_Day'].value_counts()

plt.figure(figsize=(10, 6))
sns.barplot(x=time_of_day_counts.index, y=time_of_day_counts.values, palette='viridis')
plt.title('Distribution of Fraudulent Transactions by Time of Day')
plt.xlabel('Time of Day')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
fraud_df.describe(include='object')
```

Night has the highest number of fraud transactions among other periods of the day

```python
sum_amount_by_time = fraud_df.groupby('Time_of_Day')['Amount'].sum()

plt.figure(figsize=(10, 6))
sns.barplot(x=sum_amount_by_time.index, y=sum_amount_by_time.values, palette='viridis')
plt.title('Total Amount of Fraudulent Transactions by Time of Day')
plt.xlabel('Time of Day')
plt.ylabel('Total Amount')
plt.xticks(rotation=45)
plt.show()
```

While afternoon and morning has the highest amount of money stolen in spite of the night has the most fraud transactions

```python
mean_amount_by_time = fraud_df.groupby('Time_of_Day')['Amount'].mean()

# Create the bar plot for mean
plt.figure(figsize=(10, 6))
sns.barplot(x=mean_amount_by_time.index, y=mean_amount_by_time.values, palette='viridis')
plt.title('Average Amount of Fraudulent Transactions by Time of Day')
plt.xlabel('Time of Day')
plt.ylabel('Average Amount')
plt.xticks(rotation=45)
plt.show()
```

Especially in afternoon the amount of stolen money for each transactions is higher than any other period so the thieves prefer to steal the highest amount as they can in this period. In spite of being night has the highest amount of fraud transaction, thieves don't steal the first or second amount of money in comparison with afternoon and evening

```python
pivot_table = important.pivot_table(values='Amount', index='Hours', columns='Class', aggfunc='mean')

plt.figure(figsize=(12, 6))
sns.heatmap(pivot_table, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Heatmap of Average Transaction Amount by Hour and Class')
plt.xlabel('Class (0: Non-Fraudulent, 1: Fraudulent)')
plt.ylabel('Hour of the Day')
plt.show()
```

```python
fraud_df['Amount_Bin'] = pd.cut(fraud_df['Amount'], bins)

plt.figure(figsize=(12, 6))
sns.countplot(x='Amount_Bin', data=fraud_df, color='red', label='Fraudulent', alpha=0.7)
plt.title('Fraudulent Transaction Amount Distribution by Bins')
plt.xlabel('Transaction Amount Range')
plt.ylabel('Count')
plt.legend()
plt.show()
```

The thieves prefer to steal small amount of money per transaction rather than big amounts

```python
bins = list(range(0, 2050, 50))

time_of_day_categories = fraud_df['Time_of_Day'].unique()
n_rows = len(time_of_day_categories) // 2 + len(time_of_day_categories) % 2
fig, axes = plt.subplots(nrows=n_rows, ncols=2, figsize=(14, 5 * n_rows), sharex=True, sharey=True)
axes = axes.flatten()

for i, time in enumerate(time_of_day_categories):
    ax = axes[i]
    
    time_df = fraud_df[fraud_df['Time_of_Day'] == time]
    
    sns.histplot(time_df['Amount'], bins=bins, ax=ax, color='blue', kde=False)
    ax.set_title(f'Amount Distribution for {time}')
    ax.set_xlabel('Amount')
    ax.set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```
